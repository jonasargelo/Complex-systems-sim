```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
```

```python
def visualize_grid(grid):
    """
    Parameters:
    - grid (numpy array): Grid of size L x L that represents the universe
    """

    # set color as white if site is empty; (Li,Ly) = 0
    value = 0
    masked_array = np.ma.masked_where(grid == value, grid)

    # set colormap
    cmap = mlp.cm.get_cmap('viridis')

    # plot grid
    plt.imshow(masked_array,cmap=cmap)
    plt.colorbar()
    plt.show()


N = 20
L = 20
grid = np.random.randint(N, size=(L, L))
visualize_grid(grid)
```
