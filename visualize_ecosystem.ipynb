```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
mpl.rc('font', size=4)
```

```python
def visualize_grid(grid: np.ndarray) -> None:
    """ A function to visualize universe as a grid of size L x L

    Parameters:
    ----------
    grid : numpy array
        Grid of size L x L that represents the universe

    ----------
    Note that numpy array values is mirrored horizontally in grid visualization.
    """

    # set color as white if site is empty; (Li,Ly) = 0
    value = 0
    masked_array = np.ma.masked_where(grid == value, grid)

    # set colormap
    cmap = mpl.cm.get_cmap('viridis')

    # plot grid
    plt.pcolormesh(masked_array, cmap=cmap, edgecolors='w', linewidth=0.01)
    plt.colorbar()
    ax = plt.gca()
    ax.set_aspect('equal')
    plt.show()


N = 20
L = 20
grid = np.random.randint(N, size=(L, L))
visualize_grid(grid)
```
