```python
from classes.model import Model
```

```python
import matplotlib.pyplot as plt
```

```python
def simulation(island_size, total_grid, iterations, distances):
    models = list()
    for dist in distances:
        model = Model(island_size, total_grid, iterations, distances)
        model.run()
        models.append(model)
    return models
```

```python
def distance_plotter(models, distances):
    for i in range(len(models)):
        plt.scatter(models[i].counter, distances[i])
    plt.show()
```

```python
def sa_curves(models):
    for model in models:
        model.sa_curve(model.islands[1].algorithm.current_grid)
```

```python
island_size = 50
total_grid = 20000
iterations = 1000000
```

```python
distances = [1, 5, 10]
models = simulation(distances)
```

```python
print(models)
```

```python
sa_curves(models)
```

```python
distance_plotter(models, distances)
```
