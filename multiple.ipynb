```python
from classes.model import Model
```

```python
import matplotlib.pyplot as plt
```

```python
def simulation(island_size, total_grid, iterations, distances):
    models = list()
    for dist in distances:
        model = Model(island_size, total_grid, iterations, dist)
        model.run()
        models.append(model)
    return models
```

```python
def distance_plotter(models, distances):
    for i in range(len(models)):
        plt.scatter(models[i].counter, distances[i])
    plt.show()
```

```python
def sa_curves(models):
    for model in models:
        model.sa_curve(model.islands[1].algorithm.current_grid)
```

```python
# island_size = 50
# total_grid = 20000
# iterations = 1000000
```

```python
island_size = 50
total_grid = 20000
iterations = 100000
```

```python
one_distance = int(island_size / 2) + 1
distances = [one_distance, one_distance, 10 * one_distance, 
             100 * one_distance, 1000 * one_distance]
```

```python
models = simulation(island_size, total_grid, iterations, distances)
```

```python
model = Model(island_size, total_grid, iterations, 100)
model.run()
model.convergence_plot(model.islands[0])
model.sa_curve(model.islands[0].algorithm.current_grid)
model.visualise(model.islands[0])
```

```python
sa_curves(models)
```

```python
distance_plotter(models, distances)
```
